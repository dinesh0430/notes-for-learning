```python
# Python allows you to extract the values of a collection of values in a list, tuple etc. into variables. This is called unpacking.

x,y = [1,2]
print(x)
print(y)
```

```python
# you can add an * to the variable name and the values will be assigned to the variable as a list
x,*y = [1,2,3,1,1,1,1,5]
print(x)
print(y)
print(type(y))
```

```python
# when you create a variable inside a function, that variable is local, and can only be used inside that function 
x = 'one'

def test():
    x = 'two'
    print("in func:",x)

print(x)
test()
print(x)
```

```python
# use the global keyword if you want to access a global variable inside a function
x = 'one'

def test():
    global x
    print("in func:",x)

print(x)
test()
print(x)
```

```python
# use the global keyword if you want to change a global variable inside a function

x = 'one'

def test():
    global x
    x = 23
    print("in func:",x)

print(x)
test()
print(x)
```
