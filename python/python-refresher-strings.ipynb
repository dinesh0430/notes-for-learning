# String slicing

```python
msg='abcde'
print(msg[0])
print(type(msg[0]))

print(msg[0:3]) # 0:3 means indexes 0,1,2 -- doesn't include 3
print(msg[:3])
```

```python
msg='abcde'
print(msg[0:0]) #asking for the substring that starts at index 0 and ends before index 0
print(msg[0:1])
```

```python
msg='abcde'
print(msg[3:5])
print(msg[3:])
```

```python
msg='abcde'

#  a  b  c  d  e
#  0  1  2  3  4
#  -5 -4 -3 -2 -1


print(msg[-1])
print(msg[:-1]) #print till -1 index (e), excludes index -1 (e)
print(msg[-2:]) #print from -2 index (d) to the end
print(msg[-0]) #there is no such this as -0 separately, cause -0 and 0 both mean the same
print(msg[-4:-3])
```

```python
txt = f"{'Hello':>10}"
print(txt)  # Output: '     Hello' (5 spaces before Hello)
txt = f"{'Hello':>3}"
print(txt)  # Output: 'Hello' (given 3 is less than the len of Hello (5), so no additional formatting)
txt = f"{'Hello':<10}"
print(txt)  # Output: 'Hello     ' (5 spaces after Hello)
txt = f"{'Hello':^10}"
print(txt)  # Output: '  Hello   ' (2 spaces before and 3 spaces after Hello)

num = 42
txt = f"The number is {num:05}"
print(txt)  # Output: The number is 00042

txt = f"The number is {num:5}"
print(txt)
```

```python
#Rounding while formatting

num = 3.1436
txt = f"{num:.3f}"
print(txt) 

num = 3.1435
txt = f"{num:.3f}"
print(txt) 
```

```python
num = 3.14159
txt = f"{num:10.2f}"  # Right-align with 10-character width, 2 decimal places
print(txt)  # Output: '      3.14' (6 spaces before 3.14 to make it 10 characters wide)

num = 3.14159
txt = f"{num:>10.2f}"  # Right-align with 10-character width, 2 decimal places
print(txt)  # Output: '      3.14' (6 spaces before 3.14 to make it 10 characters wide)

num = 3.14159
txt = f"{num:<10.2f}"  # Left-align with 10-character width, 2 decimal places
print(txt)  # Output: '3.14      ' (6 spaces after 3.14 to make it 10 characters wide)
```

# Methods

```python
msg='AbCdE'
print(msg.lower())
print(msg.upper())
```

```python
msg=' AbCdE'
print(len(msg)) #6

print(msg.lstrip())
print(len(msg.lstrip())) #5

print(msg.rstrip()) #doesn't work as there is no spaces to remove on the right of the string
print(len(msg.rstrip())) #6
#string removes spaces on the both the sides of the string
print(len(msg.strip())) #5
```

```python
# The strip('2x1') [ the string given is treated as a set of chars ('2','x','1') instead of a single string ] 
# strip method removes all leading and trailing characters, that are '2', 'x', or '1' from the string.
# Output:'happy' as it removes 'xx121' from the start and '122121xx' from the end.

msg = 'xx121happy122121xx'
print(msg.strip('2x1'))

msg = 'xx121happy122121Xx' # replaced one x (lowercase) with X (uppercase)
print(msg.strip('2x1')) # strip is case-sensitive

msg = 'xx121happy122H121xx' #it stops if it finds a char not present in the set of chars to be removed
print(msg.strip('2x1'))
```

```python
# split method
txt = "apple#banana#cherry#orange#happy#hermit#phone"
x = txt.split("#")
print(x)
```

```python
txt = "apple#banana#cherry#orange#happy#hermit#phone"

# setting the maxsplit parameter to 2, will return a list with 3 elements!
x = txt.split("#", 2) # (maxsplit parameter + 1) elements are returned

print(x)
```
